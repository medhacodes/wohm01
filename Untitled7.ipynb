```python

# Selecting the columns of interest
columns_to_transform = [4, 6, 7, 38]

# Extract the selected columns into a new DataFrame
new = df.iloc[1:, columns_to_transform]

# Convert the values in the DataFrame to numeric, replacing non-convertible values with NaN
new = new.apply(pd.to_numeric, errors='coerce')

# Loop through each column
for column in new:
    # Filter out zero and NaN values before applying the logarithmic transformation
    non_zero_values = new[column][(new[column] != 0) & (~np.isnan(new[column]))]
    
    if len(non_zero_values) > 0:  # Check if there are non-zero, non-NaN values
        new_distribution = np.log(non_zero_values)
        
        # Create a new figure for each column's histogram
        plt.figure(figsize=(6, 6))
        
        # Create a histogram for the current column's log-transformed values
        plt.hist(new_distribution, bins=30)
        
        plt.title(f'Histogram of Column {column} (Log Transformed)')
        plt.xlabel('Log Transformed Values')
        plt.ylabel('Frequency')
        
        plt.show()
```

```python
#removing the outliers : clipping the upper and lower bound with help of IQR and quantile 
columns_to_plot = [4, 6, 7, 11, 13, 15, 18, 20, 21, 38]

# Extract the selected columns into a new DataFrame
new = df.iloc[1:, columns_to_plot]

quantile_25 = new.quantile(0.25)

quantile_75 = new.quantile(0.75)

iqr = quantile_75 - quantile_25

lower_bound = quantile_25 - 1.5*iqr

upper_bound = quantile_75 + 1.5*iqr

clipped_data = new.clip(lower=lower_bound, upper=upper_bound,axis = 0)

```

```python

 
# Convert the values to numeric, replacing non-convertible values with NaN
new = clipped_data.apply(pd.to_numeric, errors='coerce')

# Create box plots for each column
for column in new:
    plt.figure(figsize=(6, 6))
    plt.boxplot(new[column])
    plt.title(f'Box Plot of Column {column}')
    plt.ylabel('Values')
    plt.xticks([1], [column])  # Set the x-axis label as the column name
    plt.show()
```

```python


X = df.iloc[2:,[4,6,7,11,13,15,18,20]].values


y = df.iloc[2:,22]

# Convert X to float array
X = X.astype(float)



# Convert y to float array
y = y.astype(float)



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Create a Random Forest Regressor model
#rf_model = RandomForestRegressor()

# Fit the model to the training data
#rf_model.fit(X_train, y_train)

# Predict on the testing data
#rf_y_pred = rf_model.predict(X_test)

# Create a Decision Tree Regressor model
dt_model = DecisionTreeRegressor()

# Fit the model to the training data
dt_model.fit(X_train, y_train)

# Predict on the testing data
dt_y_pred = dt_model.predict(X_test)

# Evaluate the models
#rf_mse = mean_squared_error(y_test, rf_y_pred)
#rf_r2 = r2_score(y_test, rf_y_pred)

dt_mse = mean_squared_error(y_test, dt_y_pred)
dt_r2 = r2_score(y_test, dt_y_pred)

#print("Random Forest - Mean Squared Error:", rf_mse)
#print("Random Forest - R^2 Score:", rf_r2)

print("Decision Tree - Mean Squared Error:", dt_mse)
print("Decision Tree - R^2 Score:", dt_r2)
```
